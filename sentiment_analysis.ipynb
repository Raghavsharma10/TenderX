```python
import pandas as pd
import numpy as np
path = "/Users/raghav/Downloads/IMDB Dataset.csv"
df = pd.read_csv(path)
print(df.head())
print(df.info())

```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python

   for text in df['review'].head():
       print(text)
       doc = nlp(text)
       tokens = [token.text for token in doc]
       print(f"Tokens: {tokens}")
       lemmatized_words = [token.lemma_ for token in doc]
       print(f"lemmatized: {lemmatized_words}")
```

```python
processed_reviews = []
nlp.disable_pipes("parser", "ner")
for doc in nlp.pipe(df['review'], batch_size=100,n_process=4):
    tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct and token.is_alpha]
    processed_reviews.append(' '.join(tokens))

df['processed_review'] = processed_reviews
```

```python
df['sentiment'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer_tfidf = TfidfVectorizer()
```

```python
X_tfidf = vectorizer_tfidf.fit_transform(df['processed_review'])
```

```python
from sklearn.model_selection import train_test_split

X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(X_tfidf, df['sentiment'], test_size=0.2, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python

model_nb = MultinomialNB()

model_nb.fit(X_train_tfidf, y_train_tfidf)
```

```python

y_pred_nb = model_nb.predict(X_test_tfidf)
```

```python

print(f"Accuracy: {accuracy_score(y_test_tfidf, y_pred_nb)}")
print(classification_report(y_test_tfidf, y_pred_nb))
```
